```python

import numpy as np
import pandas as pd
from math import radians
from scipy.spatial import cKDTree
import matplotlib.pyplot as plt

df = pd.read_csv("xy_data.csv")
pts = df[['x', 'y']].values
N = len(pts)

def curve_points(theta_deg, M, X, n=N, tmin=6, tmax=60):
    th = radians(theta_deg)
    t = np.linspace(tmin, tmax, n)
    exp_term = np.exp(M * np.abs(t))
    osc = np.sin(0.3 * t)
    x = t * np.cos(th) - exp_term * osc * np.sin(th) + X
    y = 42 + t * np.sin(th) + exp_term * osc * np.cos(th)
    return np.c_[x, y]

def chamfer_L1(A, B):
    TA, TB = cKDTree(A), cKDTree(B)
    da = TA.query(B, k=1)[0].mean()
    db = TB.query(A, k=1)[0].mean()
    return da + db

thetas = np.linspace(0, 50, 26)
Ms = np.linspace(-0.05, 0.05, 21)
Xs = np.linspace(0, 100, 21)

best = (1e9, None, None, None)
for th in thetas:
    for M in Ms:
        base = curve_points(th, M, 0.0)
        for X in Xs:
            cur = base.copy()
            cur[:, 0] += X
            score = chamfer_L1(cur, pts)
            if score < best[0]:
                best = (score, th, M, X)

print("Coarse best:", best)

best_score, best_th, best_M, best_X = best
fit = curve_points(best_th, best_M, best_X)
plt.figure(figsize=(6,6))
plt.scatter(pts[:,0], pts[:,1], s=4, label="Given data")
plt.plot(fit[:,0], fit[:,1], 'r', label="Fitted curve")
plt.legend()
plt.title(f"Best fit → θ={best_th:.2f}°, M={best_M:.4f}, X={best_X:.2f}")
plt.show()
```
